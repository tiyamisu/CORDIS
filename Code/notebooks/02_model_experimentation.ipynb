# CORDIS: Model Experimentation Notebook

This notebook allows interactive experimentation with the pipeline steps implemented in `src/`.

```python
import sys
import os
import pandas as pd

# Add src directory to path
sys.path.append(os.path.abspath('../'))

from src.preprocessing import load_data, clean_data, split_data
from src.features import FeaturePipeline
from src.train import train_and_tune
from src.evaluate import generate_report
from src.visualize import (
    plot_confusion_matrices,
    plot_roc_curves,
    plot_feature_importances,
    plot_model_comparison
)
```

## 1. Run Preprocessing & Feature Pipeline

```python
data_path = '../../Datasets/heart.csv'
df_raw = load_data(data_path)
df_cleaned = clean_data(df_raw, drop_duplicates=True)

X_train, X_test, y_train, y_test = split_data(df_cleaned, target_col='target', test_size=0.2, random_state=42)
```

```python
pipeline = FeaturePipeline()
X_train_trans = pipeline.fit_transform(X_train)
X_test_trans = pipeline.transform(X_test)

print(f"Transformed Training Data Shape: {X_train_trans.shape}")
X_train_trans.head()
```

## 2. Model Training & Cross-Validation tuning

```python
trained_models = train_and_tune(X_train_trans, y_train, cv_folds=5)
```

## 3. Evaluation and Results Comparison

```python
report_df = generate_report(trained_models, X_test_trans, y_test)
report_df
```

## 4. Visualizations

```python
images_dir = '../../Images/'

plot_confusion_matrices(trained_models, X_test_trans, y_test, images_dir)
plot_roc_curves(trained_models, X_test_trans, y_test, images_dir)
plot_feature_importances(trained_models, pipeline.feature_names, images_dir)
plot_model_comparison(report_df, images_dir)
```
